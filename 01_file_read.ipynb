### 진료내역
##### 주상병코드 전체 알기위한 진료내역 전체 통합 df
- 변수명 : df2
- 파일개수 : 9
- copy : df22

```python
# 진료내역 csv파일
df2 = pd.DataFrame()
for csv in glob.glob('./1일차/진료내역/*.CSV'):
    df2_temp = pd.read_csv(csv, encoding='EUC-KR')
    df2 = pd.concat([df2, df2_temp])
```

```python
temp = []
temp2 = []
for val in list(df2['주상병코드'].value_counts().index):
    if re.findall(r'^[A-Z]_', val) != []:
        temp += re.findall(r'^[A-Z]_', val)
        temp2 += [re.sub(r'^[A-Z]_', val[0] + '00', val)]
kcd_dic1 = dict(zip(temp, temp2))

temp = []
temp2 = []
for val in list(df2['부상병코드'].value_counts().index):
    if re.findall(r'^[A-Z]_', val) != []:
        temp += re.findall(r'^[A-Z]_', val)
        temp2 += [re.sub(r'^[A-Z]_', val[0] + '00', val)]
kcd_dic2 = dict(zip(temp, temp2))
```

##### 데이터 변수 정제
- 변수명 : df22
- 변환 -> 통합

```python
def Temp(csv):
    df2_temp = pd.read_csv(csv)
    
    # 변수명
    if '진료내역_2018' in csv:
        df2_temp = df2_temp.rename(columns={'가입자일련번호':'가입자 일련번호',
                                            '데이터기준일자':'데이터 기준일자'})
    
    # "달리 분류되지 않은" 코드 값 찾고 대체
    df2_temp = df2_temp.replace({'주상병코드':kcd_dic1, '부상병코드':kcd_dic2})
    
    # 상병코드 한글명 넣을 col
    df2_temp['주상병_한글명'] = df2_temp['주상병코드']
    df2_temp['부상병_한글명'] = df2_temp['부상병코드']
    
    # 데이터 정제
    df2_temp = df2_temp.replace({'시도코드':row0, '연령대코드':row2, '서식코드':row4, '진료과목코드':row6,
                                 '주상병_한글명':kcd_dic, '부상병_한글명':kcd_dic, '성별코드':{1:'남자', 2:'여자'}})
    
    # 데이터 기준일자 삭제
    df2_temp = df2_temp.drop('데이터 기준일자', axis=1)
    
    return df2_temp
```

```python
df22 = pd.DataFrame()
for csv in glob.glob('./1일차/새 폴더/*.CSV'):
    df2_temp2 = Temp(csv)
    df22 = pd.concat([df22, df2_temp2])
    print('done.............')
df22
```
